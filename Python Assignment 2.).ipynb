```python
#1.1)
class length():
    def __init__(self,d,e,f):
        self.d=d
        self.e=e
        self.f=f
        
    def cal_s(self):
        self.h=(self.d+self.e+self.f)/2
        
class triangle1(length):
    def __init__(self,*args):
        super(triangle1,self).__init__(*args)
        
    def Area1(self):
        B=(self.h*(self.h-self.d)*(self.h-self.e)*(self.h-self.f))**0.5
        print('Area of triangle is :',B)
    
        
t2=triangle1(5,5,6)
t2.cal_s()
t2.Area1()
```
